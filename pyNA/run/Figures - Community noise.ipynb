```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

from IPython.display import clear_output
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# Load data 
x = dict()
y = dict()
epnl = dict()

x['V2'] = dict()
x['Vmax'] = dict()
y['V2'] = dict()
y['Vmax'] = dict()
epnl['V2'] = dict()
epnl['Vmax'] = dict()

x['V2']['low'] = np.load('../cases/stca/output/STCB/V2/contours/x_lst_stcb_v2_low.npy')
x['Vmax']['low'] = np.load('../cases/stca/output/STCB/Vmax/contours/x_lst_stcb_vmax_low.npy')
y['V2']['low'] = np.load('../cases/stca/output/STCB/V2/contours/y_lst_stcb_v2_low.npy')
y['Vmax']['low'] = np.load('../cases/stca/output/STCB/Vmax/contours/y_lst_stcb_vmax_low.npy')
epnl['V2']['low'] = np.load('../cases/stca/output/STCB/V2/contours/epnl_stcb_v2_low.npy')
epnl['Vmax']['low'] = np.load('../cases/stca/output/STCB/Vmax/contours/epnl_stcb_vmax_low.npy')

x['V2']['high'] = np.load('../cases/stca/output/STCB/V2/contours/x_lst_stcb_v2_high.npy')
x['Vmax']['high'] = np.load('../cases/stca/output/STCB/Vmax/contours/x_lst_stcb_vmax_high.npy')
y['V2']['high'] = np.load('../cases/stca/output/STCB/V2/contours/y_lst_stcb_v2_high.npy')
y['Vmax']['high'] = np.load('../cases/stca/output/STCB/Vmax/contours/y_lst_stcb_vmax_high.npy')
epnl['V2']['high'] = np.load('../cases/stca/output/STCB/V2/contours/epnl_stcb_v2_high.npy')
epnl['Vmax']['high'] = np.load('../cases/stca/output/STCB/Vmax/contours/epnl_stcb_vmax_high.npy')

x['V2']['ptcb'] = np.load('../cases/stca/output/PTCB/V2_final/contours/x_lst.npy')
y['V2']['ptcb'] = np.load('../cases/stca/output/PTCB/V2_final/contours/y_lst.npy')
epnl['V2']['ptcb'] = np.load('../cases/stca/output/PTCB/V2_final/contours/epnl.npy')
x['Vmax']['ptcb'] = np.load('../cases/stca/output/PTCB/Vmax_final/contours/x_lst.npy')
y['Vmax']['ptcb'] = np.load('../cases/stca/output/PTCB/Vmax_final/contours/y_lst.npy')
epnl['Vmax']['ptcb'] = np.load('../cases/stca/output/PTCB/Vmax_final/contours/epnl.npy')
```

```python
# Contours
plt.figure(figsize=(15,5))
plt.style.use('../utils/plot.mplstyle')
plt.contour(x['V2']['high'], y['V2']['high'], epnl['V2']['high'], levels=[80], linestyles='-', colors='tab:blue')
plt.contour(x['Vmax']['high'], y['V2']['high'], epnl['Vmax']['high'], levels=[80], linestyles='--', colors='tab:blue')

plt.contour(x['V2']['low'], y['V2']['low'], epnl['V2']['low'], levels=[80], colors='tab:orange')
plt.contour(x['Vmax']['low'], y['V2']['low'], epnl['Vmax']['low'], levels=[80], linestyles='--', colors='tab:orange')

plt.contour(x['V2']['ptcb'], y['V2']['ptcb'], epnl['V2']['ptcb'], levels=[80], colors='tab:green')
plt.contour(x['Vmax']['ptcb'], y['V2']['ptcb'], epnl['Vmax']['ptcb'], levels=[80], linestyles='--', colors='tab:green')
```

```python
plt.figure(figsize=(8,3))
plt.style.use('../utils/plot.mplstyle')
plt.title('STCB ($z_{cb} \geq 260m, 250 kts$)')

c = plt.contour(x['V2']['high']/1e3, y['V2']['high']/1e3, epnl['V2']['high'], levels=[60, 70, 80, 90], colors='k')

plt.clabel(c, inline=1, fontsize=10, fmt='%1.0f')
```
