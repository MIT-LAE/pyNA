```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import clear_output
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# Inputs control schedules
x = np.linspace(0, 100, 88)
y1 = np.ones(88); y1[41:] = 0.5*np.ones(47)
y2 = np.ones(88); y2[30:52] = np.linspace(1, 0.5, 22); y2[52:] = 0.5*np.ones(36)
y3 = np.ones(88); y3[30:52] = np.linspace(1, 0.5, 22)+0.05*np.sin(2*np.pi*np.linspace(0, 1, 22)); y3[52:] = 0.5*np.ones(36)
y4 = np.ones(88); y4[30:52] = np.linspace(1, 0.5, 22)+0.05*np.sin(2*np.pi*np.linspace(0, 2, 22)); y4[52:] = 0.5*np.ones(36)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,4))
plt.style.use('../utils/plot.mplstyle')

ax[0].plot(x, np.mean(y1[30:52])*np.ones(np.size(x)), color='grey', alpha=0.5)
ax[0].plot(x, y1)
ax[0].plot(x, y2)
ax[0].plot(x, y3)
ax[0].plot(x, y4)
ax[0].set_xlabel('t_s [s]')
ax[0].set_ylabel('$TS$')
ax[0].set_ylim([0, 1.05])

exp = 3
ax[1].plot(x, np.mean(y1[30:52]**exp)*np.ones(np.size(x)), color='grey', alpha=0.5)
ax[1].plot(x, y1**exp)
ax[1].plot(x, y2**exp)
ax[1].plot(x, y3**exp)
ax[1].plot(x, y4**exp)
ax[1].set_xlabel('t_s [s]')
ax[1].set_ylabel('$TS^3$')
ax[1].set_ylim([0, 1.05])
```

```python
# Print areas
print("Area A1: ", np.trapz(y1, x))
print("Area A2: ", np.trapz(y2, x))
print("Area A3: ", np.trapz(y3, x))
print("Area A4: ", np.trapz(y4, x))
```

```python
# Print areas
print("Area A1: ", np.mean(y1))
print("Area A2: ", np.mean(y2))
print("Area A3: ", np.mean(y3))
print("Area A4: ", np.mean(y4))
```

```python
# Print new objective function
print(sum(y1[30:52]**exp/np.size(y1[30:52])))
print(sum(y2[30:52]**exp/np.size(y2[30:52])))
print(sum(y3[30:52]**exp/np.size(y3[30:52])))
print(sum(y4[30:52]**exp/np.size(y4[30:52])))
```
